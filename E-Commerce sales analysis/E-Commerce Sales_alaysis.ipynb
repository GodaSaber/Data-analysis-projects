```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

```python
df["Year"]=df["Order_Date"].dt.year
df["Month"] = df["Order_Date"].dt.month_name()
df["Day"]=df["Order_Date"].dt.day_name()
```

```python
df.head()
```

```python
ym_groupby_order_id=df.groupby(["Year","Month"])["Order_ID"].count().reset_index()
```

```python
ym_groupby_order_id.head()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=ym_groupby_order_id,x="Year",y="Order_ID")
plt.title("Number of Orders per Month")
plt.xticks(rotation=45)
plt.show()
```

```python
yc=df["Year"].value_counts()
yc
```

```python
plt.pie(yc.values,labels=yc.index,autopct='%1.1f%%',explode = (.1, .1, .1), shadow=True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=ym_groupby_order_id,x="Month",y="Order_ID")
plt.title("Number of Orders per Month")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=ym_groupby_order_id, x="Month", y="Order_ID", hue="Year", marker="o",)
plt.title("Number of Orders per Month")
plt.xticks(rotation=45)
plt.show()
```
